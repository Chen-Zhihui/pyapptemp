```python
import os
import xml.etree.ElementTree as ET

class ScenceInfo(object):

    def __init__(self):
        self.scence_id = None 
        self.satellite = None
        self.sensor = None
        self.rec_date_utc  = None
        self.clat = None 
        self.clon = None
        self.ne_lat = None 
        self.ne_lon = None
        self.nw_lat = None 
        self.nw_lon = None
        self.sw_lat = None
        self.sw_lon = None
        self.se_lat = None
        self.se_lon = None
        self.cloud_pre = None 

    def __str__(self):
        return f"""
        scence_id={self.scence_id}, satellite={self.satellite}, sensor={self.sensor}
        rec_date_utc={self.rec_date_utc}
        center: clat={self.clat}, clon={self.clon}
        ne : ne_lat={self.ne_lat}, ne_lon={self.ne_lon}
        nw : nw_lat={self.nw_lat}, nw_lon={self.nw_lon}
        sw : sw_lat={self.sw_lat}, sw_lon={self.sw_lon}
        se : se_lat={self.se_lat}, se_lon={self.se_lon}
        cloud_pre={self.cloud_pre}
        """

def get_scence_from_21_TRIPLESAT_1(filepath):
    tree = ET.parse(filepath)
    info = ScenceInfo()
    root = tree.getroot()
    for ele in root[0]:
        if ele.tag == "Scene_ID":
            info.scence_id = ele.text
        elif ele.tag == "Satellite_Name":
            info.satellite = ele.text
        elif ele.tag == "Sensor_Name":
            info.sensor = ele.text
        elif ele.tag == "End_Time":
            info.rec_date_utc = ele.text
        elif ele.tag == "Central_Lat":
            info.clat = ele.text
        elif ele.tag == "Central_Lon":
            info.clon = ele.text
        elif ele.tag == "UR_Lat":
            info.ne_lat = ele.text
        elif ele.tag == "UR_Lon":
            info.ne_lon = ele.text
        elif ele.tag == "UL_Lat":
            info.nw_lat = ele.text
        elif ele.tag == "UL_Lon":
            info.nw_lon = ele.text
        elif ele.tag == "LL_Lat":
            info.sw_lat = ele.text
        elif ele.tag == "LL_Lon":
            info.sw_lon = ele.text
        elif ele.tag == "LR_Lat":
            info.se_lat = ele.text
        elif ele.tag == "LR_Lon":
            info.se_lon = ele.text
        elif ele.tag == "Cloud_Cover":
            info.cloud_pre = ele.text
        else:
            pass
    return info

def get_scence_from_21_BJ3A1_PMS1(filepath):
    tree = ET.parse(filepath)
    root = tree.getroot()
    info = ScenceInfo()
    for ele in root[0]:
        if ele.tag == "Scene_ID":
            info.scence_id = ele.text
        elif ele.tag == "Satellite_Name":
            info.satellite = ele.text
        elif ele.tag == "Camera_Name":
            info.sensor = ele.text
        elif ele.tag == "Acquisition_Time_UTC":
            info.rec_date_utc = ele.text
        elif ele.tag == "Central_Lat":
            info.clat = ele.text
        elif ele.tag == "Central_Lon":
            info.clon = ele.text
        elif ele.tag == "UR_Lat":
            info.ne_lat = ele.text
        elif ele.tag == "UR_Lon":
            info.ne_lon = ele.text
        elif ele.tag == "UL_Lat":
            info.nw_lat = ele.text
        elif ele.tag == "UL_Lon":
            info.nw_lon = ele.text
        elif ele.tag == "LL_Lat":
            info.sw_lat = ele.text
        elif ele.tag == "LL_Lon":
            info.sw_lon = ele.text
        elif ele.tag == "LR_Lat":
            info.se_lat = ele.text
        elif ele.tag == "LR_Lon":
            info.se_lon = ele.text
        elif ele.tag == "Cloud_Cover":
            info.cloud_pre = ele.text
        else:
            pass
    return info


def get_scence_from_ziyuan_SV1_03(filepath):
    tree = ET.parse(filepath)
    root = tree.getroot()
    info = ScenceInfo()
    info.scence_id = os.path.basename(filepath)
    for ele in root:
        if ele.tag == "SatelliteID":
            info.satellite = ele.text
        elif ele.tag == "SensorID":
            info.sensor = ele.text
        elif ele.tag == "EndTime":
            info.rec_date_utc = ele.text
        elif ele.tag == "CenterLatitude":
            info.clat = ele.text
        elif ele.tag == "CenterLongitude":
            info.clon = ele.text
        elif ele.tag == "TopRightLatitude":
            info.ne_lat = ele.text
        elif ele.tag == "TopRightLongitude":
            info.ne_lon = ele.text
        elif ele.tag == "TopLeftLatitude":
            info.nw_lat = ele.text
        elif ele.tag == "TopLeftLongitude":
            info.nw_lon = ele.text
        elif ele.tag == "BottomLeftLatitude":
            info.sw_lat = ele.text
        elif ele.tag == "BottomLeftLongitude":
            info.sw_lon = ele.text
        elif ele.tag == "BottomRightLatitude":
            info.se_lat = ele.text
        elif ele.tag == "BottomRightLongitude":
            info.se_lon = ele.text
        elif ele.tag == "CloudPercent":
            info.cloud_pre = ele.text
        else:
            pass
    return info

SATELLITE_INFOS = {
    # Obit
    "TRIPLESAT_1"   : {"reader" : get_scence_from_21_TRIPLESAT_1},
    "BJ3A1"         : {"reader" : get_scence_from_21_BJ3A1_PMS1},
    # ZiYuan
    "SV1-03"        : {"reader" : get_scence_from_ziyuan_SV1_03}
}

def satelliet_info(filepath):    
    basename = os.path.basename(filepath)
    for name in SATELLITE_INFOS:
        if basename.startswith(name):
            return SATELLITE_INFOS[name]
    return None

def info_from_xml(filepath):
    # 每个星有自己的XML格式，文件名称中的第一个字体段即星名
    sinfo = satelliet_info(filepath)
    if sinfo is not None and sinfo["reader"] is not None:
        return sinfo["reader"](filepath)
    else :
        return None





```
